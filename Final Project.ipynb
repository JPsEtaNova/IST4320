```python
import tkinter as tk
import os
import re

email_set = set()

def submit_information():
  name = user_name_entry.get()
  email = user_email_entry.get()
  phone_number = user_phone_number_entry.get()

  # Validate input
  if not name:
    error_label["text"] = "Please enter your name."
    return
  if not email:
    error_label["text"] = "Please enter your email address."
    return
  if not re.match(r"\d{3}-\d{3}-\d{4}", phone_number):
    error_label["text"] = "Please enter a valid phone number (XXX-XXX-XXXX)."
    return

  # Check if email already exists in the set
  if email in email_set:
    error_label["text"] = "Email address already exists!"
    return

  # Otherwise, add email to the set
  email_set.add(email)

  # Save information to a file
  filename = "user_data.txt"
  with open(filename, "a") as file:
    file.write(f"{name},{email},{phone_number}\n")

  # Clear entry fields
  user_name_entry.delete(0, tk.END)
  user_email_entry.delete(0, tk.END)
  user_phone_number_entry.delete(0, tk.END)

  # Display success message
  error_label["text"] = "Information saved successfully!"

root = tk.Tk()
root.title("User Information Form")

user_name_label = tk.Label(root, text="Name:")
user_name_label.grid(row=0, column=0)

user_name_entry = tk.Entry(root)
user_name_entry.grid(row=0, column=1)

user_email_label = tk.Label(root, text="Email:")
user_email_label.grid(row=1, column=0)

user_email_entry = tk.Entry(root)
user_email_entry.grid(row=1, column=1)

user_phone_number_label = tk.Label(root, text="Phone Number:")
user_phone_number_label.grid(row=2, column=0)

user_phone_number_entry = tk.Entry(root)
user_phone_number_entry.grid(row=2, column=1)

submit_button = tk.Button(root, text="Submit", command=submit_information)
submit_button.grid(row=3, column=0, columnspan=2)

error_label = tk.Label(root, text="", fg="red")
error_label.grid(row=4, column=0, columnspan=2)


# Check if data file exists and create it if necessary
filename = "user_data.txt"
if not os.path.exists(filename):
    with open(filename, "w") as file:
        file.write("")

root.mainloop()
```
